# Changes in the calculated values of KL divergence when less samples are used.

I don't calculate the value of the KL divergence for ADMC because it has not converged at 28000th iterations.

Note that the nonzero starting points are the vectors from the 12 sets: 
(1) k \times I_{50}
(2) k \times companion

There is a rationale for choosing groups of 10 points in each of the 12 sets above. (as if you have run fewer number of chains from starting positions over [-3,3]^{50}).

Basic BPS; measure the KL divergence at t=28000.

```
dim = 50

KL_divergence_with_SN = lambda mu_1, Sigma_1: 0.5*(np.trace(np.linalg.inv(Sigma_1)) + mu_1.dot(np.linalg.inv(Sigma_1)).dot(mu_1) - 
                                           len(mu_1) + math.log(np.linalg.det(Sigma_1)))


samples_at_28000 = np.zeros((600,50))
reduced_KLs_list = []


for i in range(600):
  path_to_load_BPS = 'processed_BPS_data_SFORD/processed_BPS_chain_' + str(i) + '_x_array.npz'
  previous_data = np.load(path_to_load_BPS)
  previous_chain = previous_data['x_array']

  samples_at_28000[i] = previous_chain[280]




```

```
for j in [10,20,30,40,50]:

  consecutives = np.arange(0,j)
  selected_indices = np.arange(0,j)
  
  for k in range(1,12):
    new_indices = k*50 + consecutives
    selected_indices = np.append(selected_indices, new_indices)
  
  reduced_samples = samples_at_28000[selected_indices]

  # debugging
  if reduced_samples.shape == (12*j, dim):
    print('shape of reduced_samples is correct: j: {}      shape: {}'.format(j, reduced_samples.shape))

  mean = np.mean(reduced_samples, axis=0)
  cov_mat = np.cov(reduced_samples, rowvar=False)

  shrinked_KL = KL_divergence_with_SN(mean, cov_mat)

  reduced_KLs_list.append(shrinked_KL)

```

Basic BPS

```
# The following function allows me to find the computational time required to achieve 
# a certain number of markov chain iterations.

def CPUTime_and_MarkovChainTime(cpu_list, stride_times_list, markov_time):
  
  indicator_list = [1 if x < markov_time else 0 for x in stride_times_list]

  return cpu_list[np.sum(np.array(indicator_list))]
```

```
indices = [(0,'25508'), (1, '2010'), (2, 'EFORD'), (3, 'SFORD'), (4, 'BFORD'), (5, 'TFORD')]

computational_times_list_BPS = []

for i in indices:
  for j in range(20):
    path_to_load_cpu = 'BPS_data_' + i[1] + '/BPS_chain_' + str(i[0]*100 + j) + '_CPUtime_list.npz'
    previous_data_cpu = np.load(path_to_load_cpu)
    CPUtimes = previous_data_cpu['CPUtime_list']

    path_to_load_strides = 'BPS_data_' + i[1] + '/BPS_chain_' + str(i[0]*100 + j) + '_stride_times.npz'
    previous_data_strides = np.load(path_to_load_strides)
    stride_times = previous_data_strides['stride_times']

    computational_time = CPUTime_and_MarkovChainTime(CPUtimes, stride_times, 28000)
    computational_times_list_BPS.append(computational_time)


BPS_cpu_time_average = np.mean(np.array(computational_times_list_BPS))
```
